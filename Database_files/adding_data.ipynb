```python
import pandas as pd
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, String, Date, Float, DateTime

# Database connection settings
db_config = {
    "dbname": "postgres",
    "user": "postgres",
    "password": "finalproject",
    "host": "database-1.cxqtmtcu0z5m.us-east-1.rds.amazonaws.com",
    "port": "5432"  # Default is 5432 for PostgreSQL
}

# Function to create a table
def create_table(engine, table_name, columns):
    metadata = MetaData()
    table = Table(table_name, metadata)
    for col_name, col_type in columns.items():
        table.append_column(Column(col_name, col_type))
    metadata.create_all(engine)

# Function to insert data into the database
def insert_data(df, table_name, engine):
    df.to_sql(table_name, engine, if_exists='append', index=False)

# Connect to the PostgreSQL database
engine = create_engine(f"postgresql://{db_config['user']}:{db_config['password']}@{db_config['host']}:{db_config['port']}/{db_config['dbname']}")

# Define the table schemas and create the tables
Population_Columns = {
    'state': String,
    '2016': Float,
    '2017': Float,
    '2018': Float,
    '2019': Float,
    '2020': Float,
    '2021': Float,
}
create_table(engine, 'education', Population_Columns)

#Import and insert charging station data
df_education = pd.read_csv("bachelor_degree_fix.csv")
# Define the numeric columns that need commas removed
numeric_columns = [
    '2016',
    '2017',
    '2018',
    '2019',
    '2020',
    '2021'
]

# Remove commas from the numeric columns and convert them to integers
for col in numeric_columns:
    column = str(col)
    if df_education[column].dtype == object:
        df_education[column] = df_education[column].str.replace(',', '').astype(float)
    df_education[column] = (df_education[column] / 100).round(4)



insert_data(df_education, 'education', engine)
```

CREATE TABLE ev_registration_gigaa AS
SELECT 
    ev.state,
    ev.year,
    ev.electric_ev,
	ev.plug_in_hybrid_electric_phev,
    cs.charging_station_count,
    ni.income,
    ne.education,
    np.population
FROM ev_registration ev
LEFT JOIN (
    SELECT state, EXTRACT(year FROM open_date)::integer AS year, COUNT(*) AS charging_station_count
    FROM charging_stations
    GROUP BY state, EXTRACT(year FROM open_date)
) cs ON ev.state = cs.state AND ev.year = cs.year
LEFT JOIN new_income ni ON ev.state = ni.state AND ev.year = ni.year::integer
LEFT JOIN new_education ne ON ev.state = ne.state AND ev.year = ne.year::integer
LEFT JOIN new_population np ON ev.state = np.state AND ev.year = np.year::integer
ORDER BY ev.state, ev.year;
